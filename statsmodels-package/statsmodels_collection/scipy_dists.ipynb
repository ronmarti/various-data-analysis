## Expected value of function | SciPy
The main takeaway here is: remember to normalize by the interval's probability content (cdf(ub) - cdf(lb)).

```python
import matplotlib.pyplot as plt
from scipy.stats import lognorm, expon, gamma, norm
import numpy as np


%matplotlib widget
```

```python
mu = 4.29
s = np.sqrt(0.9)

# freeze distribution parameters into the random variable rv
rv = gamma(a=4, loc=2.29, scale=.7)

print(f'cdf -inf:4.29: {rv.cdf(4.29)}')
x = np.linspace(0, 15, 1000)

fig, ax = plt.subplots(1, 1)
ax.plot(x, rv.pdf(x), 'r-', lw=1, alpha=0.6, label='gamma pdf')
ax.plot(x, norm.pdf(x, loc=4.29, scale=np.sqrt(5)), 'b-', lw=1, alpha=0.6, label='norm pdf')
ax.legend()
ax.grid()
```

```python
lb=0
ub=4.29
exp_val = rv.expect(lb=lb, ub=ub) / (rv.cdf(ub)-rv.cdf(lb))
print(f'expected value of interval {lb}:{ub} = {exp_val}')

print(f'Mean: {rv.mean()}')
```
