```python
import os
print(f'Current working dir:\t {os.getcwd()}')
from pathlib import Path

from datetime import datetime
# from datetime import timedelta

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

import torch

%matplotlib widget
```

```python
x = np.linspace(2., 1e7, 100)
w1 = 6
w2 = w1 + 1
b1 = 20
b2 = b1

d1 = b1**w1
d2 = b2**w2
Nc1 = x - x*((d1-1)/d1)**(x-1)
Nc2 = x - x*((d2-1)/d2)**(x-1)
ratio = Nc1/Nc2
# print(f'Nc1 = {Nc1:.2f}\nNc2 = {Nc2:.2f}')
# print(f'Nc1/Nc2 = {Nc1/Nc2:.2f}')

plt.plot(x, ratio)
```
